```python
import numpy as np
from numpy.linalg import inv
import csv
import math
import matplotlib.pyplot as plt
```

```python
#Datasets
DS1_mean0, DS1_mean1, DS1_cov = [],[],[]
DS2_mean0, DS2_mean1, DS2_cov = [],[],[]
DS3_mean0, DS3_mean1, DS3_cov = [],[],[]
```

```python
#############################Importing Data from Files###############################

#Dataset 1
with open("./Datasets/DS2_c1_m1.txt", 'rb') as csv_mean0_file:
    mean0_data = csv.reader(csv_mean0_file, delimiter = ',')
    for row in mean0_data:
        DS1_mean0.append(row)
    del(row)

with open("./Datasets/DS2_c2_m1.txt", 'rb') as csv_mean1_file:
    mean1_data = csv.reader(csv_mean1_file, delimiter = ',')
    for row in mean1_data:
        DS1_mean1.append(row)
    del(row)

with open("./Datasets/DS2_Cov1.txt", 'rb') as csv_cov_file:
    cov_data = csv.reader(csv_cov_file, delimiter = ',')
    for row in cov_data:
        DS1_cov.append(row)
    del(row)
    
DS1_mean1 = np.delete(DS1_mean1, len(DS1_mean1[0])-1, 1)
DS1_mean1 = (DS1_mean1[0]).astype(np.float)
DS1_mean0= np.delete(DS1_mean0, len(DS1_mean0[0])-1, 1)
DS1_mean0 = (DS1_mean0[0]).astype(np.float)
DS1_cov =  (np.delete(DS1_cov, len(DS1_cov[0])-1, 1)).astype(np.float)
```

```python
#Dataset 2
with open("./Datasets/DS2_c1_m2.txt", 'rb') as csv_mean0_file:
    mean0_data = csv.reader(csv_mean0_file, delimiter = ',')
    for row in mean0_data:
        DS2_mean0.append(row)
    del(row)

with open("./Datasets/DS2_c2_m2.txt", 'rb') as csv_mean1_file:
    mean1_data = csv.reader(csv_mean1_file, delimiter = ',')
    for row in mean1_data:
        DS2_mean1.append(row)
    del(row)

with open("./Datasets/DS2_Cov2.txt", 'rb') as csv_cov_file:
    cov_data = csv.reader(csv_cov_file, delimiter = ',')
    for row in cov_data:
        DS2_cov.append(row)
    del(row)
    
DS2_mean1 = np.delete(DS2_mean1, len(DS2_mean1[0])-1, 1)
DS2_mean1 = (DS2_mean1[0]).astype(np.float)
DS2_mean0= np.delete(DS2_mean0, len(DS2_mean0[0])-1, 1)
DS2_mean0 = (DS2_mean0[0]).astype(np.float)
DS2_cov =  (np.delete(DS2_cov, len(DS2_cov[0])-1, 1)).astype(np.float)
```

```python
#Dataset 3
with open("./Datasets/DS2_c1_m3.txt", 'rb') as csv_mean0_file:
    mean0_data = csv.reader(csv_mean0_file, delimiter = ',')
    for row in mean0_data:
        DS3_mean0.append(row)
    del(row)

with open("./Datasets/DS2_c2_m3.txt", 'rb') as csv_mean1_file:
    mean1_data = csv.reader(csv_mean1_file, delimiter = ',')
    for row in mean1_data:
        DS3_mean1.append(row)
    del(row)

with open("./Datasets/DS2_Cov3.txt", 'rb') as csv_cov_file:
    cov_data = csv.reader(csv_cov_file, delimiter = ',')
    for row in cov_data:
        DS3_cov.append(row)
    del(row)
    
DS3_mean1 = np.delete(DS3_mean1, len(DS3_mean1[0])-1, 1)
DS3_mean1 = (DS3_mean1[0]).astype(np.float)
DS3_mean0= np.delete(DS3_mean0, len(DS3_mean0[0])-1, 1)
DS3_mean0 = (DS3_mean0[0]).astype(np.float)
DS3_cov =  (np.delete(DS3_cov, len(DS3_cov[0])-1, 1)).astype(np.float)
```

```python
#Dataset1

#Generating data for class 0
DS1_class0 = np.random.multivariate_normal(DS1_mean0,DS1_cov, int(2000*0.1))

#Generating data for class 1
DS1_class1 = np.random.multivariate_normal(DS1_mean1,DS1_cov, int(2000*0.1))

#Dataset2

#Generating data for class 0
DS2_class0 = np.random.multivariate_normal(DS2_mean0,DS2_cov, int(2000*0.42))

#Generating data for class 1
DS2_class1 = np.random.multivariate_normal(DS2_mean1,DS2_cov, int(2000*0.42))

#Dataset3

#Generating data for class 0
DS3_class0 = np.random.multivariate_normal(DS3_mean0,DS2_cov, int(2000*0.48))

#Generating data for class 1
DS3_class1 = np.random.multivariate_normal(DS3_mean1,DS3_cov, int(2000*0.48))
```

```python
##################### Recalculating Mean and Covariance #######################

#Dataset 1
DS1_mean0_new, DS1_mean1_new = [],[]
for i in range(len(DS1_class0[0])):
    DS1_mean0_new.append(np.mean(DS1_class0[:,i]))
    DS1_mean1_new.append(np.mean(DS1_class1[:,i]))

#Defining mean and Covariance    
DS1_mean0 = np.array(DS1_mean0_new, dtype = 'float')
DS1_mean1 = np.array(DS1_mean1_new, dtype = 'float')
DS1_cov = np.cov(np.append(DS1_class0, DS1_class1, axis = 0).T)


#Dataset 2
DS2_mean0_new, DS2_mean1_new = [],[]
for i in range(len(DS2_class0[0])):
    DS2_mean0_new.append(np.mean(DS2_class0[:,i]))
    DS2_mean1_new.append(np.mean(DS2_class1[:,i]))

#Defining mean and Covariance    
DS2_mean0 = np.array(DS2_mean0_new, dtype = 'float')
DS2_mean1 = np.array(DS2_mean1_new, dtype = 'float')
DS2_cov = np.cov(np.append(DS2_class0, DS2_class1, axis = 0).T)


#Dataset 3
DS3_mean0_new, DS3_mean1_new = [],[]
for i in range(len(DS1_class0[0])):
    DS3_mean0_new.append(np.mean(DS3_class0[:,i]))
    DS3_mean1_new.append(np.mean(DS3_class1[:,i]))

#Defining mean and Covariance    
DS3_mean0 = np.array(DS3_mean0_new, dtype = 'float')
DS3_mean1 = np.array(DS3_mean1_new, dtype = 'float')
DS3_cov = np.cov(np.append(DS3_class0, DS3_class1, axis = 0).T)
```

```python
#################  Labelling Class 0  ########################

#Dataset 1
DS1_class0 = np.hstack((DS1_class0, np.zeros((len(DS1_class0) ,1))))

#Dataset 2
DS2_class0 = np.hstack((DS2_class0, np.zeros((len(DS2_class0) ,1))))

#Dataset 3
DS3_class0 = np.hstack((DS3_class0, np.zeros((len(DS3_class0) ,1))))

#########################  Labelling Class 1  ##########################

#Dataset 1
DS1_class1 = np.hstack((DS1_class1, np.ones((len(DS1_class1),1))))

#Dataset 2
DS2_class1 = np.hstack((DS2_class1, np.ones((len(DS2_class1) ,1))))

#Dataset 3
DS3_class1 = np.hstack((DS3_class1, np.ones((len(DS3_class1) ,1))))
```

```python
#Combining data of both classes from all datasets
class0 = np.append(DS1_class0, (np.append(DS2_class0, DS3_class0, axis = 0)) , axis = 0)
class1 = np.append(DS1_class1, (np.append(DS2_class1, DS3_class1, axis = 0)) , axis = 0)
raw_data = np.append(class0, class1, axis = 0)
```

```python
#Creating training set
X_train = np.append(class0[:int(len(class0)*0.7)], class1[:int(len(class1)*0.7)], axis = 0) 
np.random.shuffle(X_train)
Y_train = X_train[:,-1]
X_train = np.delete(X_train, -1, 1)

#Creating test set
X_test = np.append(class0[int(len(class0)*0.7):], class1[int(len(class1)*0.7):], axis = 0) 
np.random.shuffle(X_test)
Y_test = X_test[:,-1]
X_test = np.delete(X_test, -1, 1)
DS2 = X_test
```

```python
#Writing Dataset to csv_file
file_name_str = "Dataset_DS2.csv"
myFile = open(file_name_str, 'w') 
with myFile:
    writer = csv.writer(myFile)
    writer.writerows(X_test)
```

```python
##################### Recalculating Mean and Covariance of new Dataset #######################
class0 = np.delete(class0, -1, 1)
class1 = np.delete(class1, -1, 1)


mean0_new, mean1_new = [],[]
for i in range(len(class0[0])):
    mean0_new.append(np.mean(class0[:int(len(class0)*0.7):,i]))
    mean1_new.append(np.mean(class1[:int(len(class0)*0.7):,i]))

#Defining mean and Covariance    
mean0 = np.array(mean0_new, dtype = 'float')
mean1 = np.array(mean1_new, dtype = 'float')
cov = np.cov(np.append(class0[:int(len(class0)*0.7)], class1[:int(len(class0)*0.7)] , axis = 0).T)
```

```python
################### Calculating parameters  ########################
prob_C1 = float(np.count_nonzero(Y_train))/float(len(X_train))
prob_C2 = 1 - prob_C1

params = np.dot(inv(cov),(mean0 - mean1))
param0 = -0.5* np.dot(np.dot((mean0.T),  inv(cov)), mean0)  + 0.5* np.dot(np.dot((mean1.T),  inv(cov)), mean1) + math.log(prob_C1/prob_C2)
```

```python
#Writing params to csv_file
file_name_str = "Params_DS2.csv"
myFile = open(file_name_str, 'w') 
with myFile:
    writer = csv.writer(myFile)
    writer.writerow(params)
```

```python
#Finding probability of data being in class 0 for training set and adding to list
train_prob_c1_x = []

for row in X_train:
    pred = np.dot(params.T, row) + param0
    train_prob_c1_x.append(1/(1 + math.exp(-pred)))
```

```python
#Lists used to keep count of data in respective classes
train_count_C1, train_count_C2 = [],[]

for value in train_prob_c1_x:
    if(value >= 0.5):
        train_count_C1.append(value)
    else:
        train_count_C2.append(value)
```

```python
#Finding probability of data being in class 0 for test set and adding to list
test_prob_c1_x = []

for row in X_test:
    pred = np.dot(params.T, row) + param0
    test_prob_c1_x.append(1/(1 + math.exp(-pred)))
```

```python
#Lists used to keep count of test data in respective classes
test_count_C1, test_count_C2 = [],[]
pred_class = []
test_TP, test_TN, test_FP, test_FN = 0,0,0,0

for value in test_prob_c1_x:
    if(value >= 0.5):
        test_count_C1.append(value)
        pred_class.append(0)
    else:
        test_count_C2.append(value)
        pred_class.append(1)
```

```python
#Calculating TP,TN, FP, FN
for i in range(len(pred_class)):
    if(pred_class[i] == Y_test[i]):
        if(pred_class[i] == 0):
            test_TP += 1
        if(pred_class[i] == 1):
            test_TN += 1    
    else:
        if(pred_class[i] == 0):
            test_FP += 1
        if(pred_class[i] == 1):
            test_FN += 1
```

```python
accuracy = (float(test_TP + test_TN))/(float(len(pred_class)))
precision =  (float(test_TP))/(float(test_TP + test_FP))
recall = (float(test_TP))/(float(test_TP + test_FN))
f_score = (2*precision*recall)/(precision + recall)
```

```python
print(accuracy)
print(precision)
print(recall)
print(f_score)
```

```python
################################## Question3:  K Nearest ###############################################
knn_pred_class = []
value_k = 5
```

```python
for test in range(len(X_test)):
    min_distance_arr, min_class_arr = [],[]
    for k in range(value_k):
        min_distance = 10000
        min_class = 0
        for train in range(len(X_train)):
            distance = 0
            for i in range(len(X_train[train])):
                distance += (X_train[train][i] - X_test[test][i])**2
            if(len(min_distance_arr) == 0):
                if(distance<min_distance):
                    min_distance = distance
                    min_class = Y_train[train]
            else:
                if(distance<min_distance and distance > min_distance_arr[len(min_distance_arr)-1]):
                    min_distance = distance
                    min_class = Y_train[train]
                    
        if(len(min_distance_arr) < value_k):
            min_distance_arr.append(min_distance)
            min_class_arr.append(min_class)
    
    string = "Iteration: " + str(test) + " Array: " + str(min_distance_arr)
    print(string)
    if(min_class_arr.count(0) > min_class_arr.count(1)):
        knn_pred_class.append(0)
        print(min_class_arr.count(0))
    else:
        knn_pred_class.append(1)
        print(min_class_arr.count(1))
```

```python
knn_test_TP, knn_test_TN, knn_test_FP, knn_test_FN = 0,0,0,0

#Calculating TP,TN, FP, FN
for i in range(len(knn_pred_class)):
    if(knn_pred_class[i] == Y_test[i]):
        if(knn_pred_class[i] == 0):
            knn_test_TP += 1
        if(knn_pred_class[i] == 1):
            knn_test_TN += 1    
    else:
        if(knn_pred_class[i] == 0):
            knn_test_FP += 1
        if(knn_pred_class[i] == 1):
            knn_test_FN += 1
```

```python
knn_accuracy = (float(knn_test_TP + knn_test_TN))/(float(len(knn_pred_class)))
knn_precision =  (float(knn_test_TP))/(float(knn_test_TP + knn_test_FP))
knn_recall = (float(knn_test_TP))/(float(knn_test_TP + knn_test_FN))
knn_f_score = (2*knn_precision*knn_recall)/(knn_precision + knn_recall)
```
